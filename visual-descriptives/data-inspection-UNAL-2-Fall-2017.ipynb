## Construct panel data for UNAL 2 with colors 

```python
import pandas as pd
import seaborn as sns

df_45 = pd.read_csv("../data/Unal-2-Phase-4-_Unal-2-Phase-5–raw-results.csv")
df_34 = pd.read_csv("../data/Unal-2-Phase-3-_Unal-2-Phase-4–raw-results.csv")
df_23 = pd.read_csv("../data/Unal-2-Phase-2-_Unal-2-Phase-3–raw-results.csv")
df_12 = pd.read_csv("../data/Unal-2-Phase-1-_Unal-2-Phase-2–raw-results.csv")
df_12['stage'] = 1
df_23['stage'] = 2
df_34['stage'] = 3
df_45['stage'] = 4
```

```python
df_color = pd.concat([df_12, df_23, df_34, df_45], ignore_index=True)
df_color
```

```python
ax = sns.lineplot(data=df_color, x='stage', y='ssim', hue='comparison_type')
ax.legend(loc='center right', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

## Individual Exploration for UNAL 0

```python
df_color['link1_change'] = df_color['link1'].str.split('/', expand=True)[4]
df_color['link2_change'] = df_color['link2'].str.split('/', expand=True)[4]
df_color
```

```python
df_color_exp = df_color[df_color['link1_change'] == df_color['link2_change']]
df_color_exp['ssim_exp'] = 100 - df_color_exp['ssim']
```

```python
ax = sns.lineplot(data=df_color_exp, x='stage', y='ssim_exp', hue='comparison_type')
ax.legend(loc='center right', bbox_to_anchor=(1.25, 0.5), ncol=1)
```
