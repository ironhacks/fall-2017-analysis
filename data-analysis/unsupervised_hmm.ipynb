```python
from hmmlearn import hmm
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from sklearn import datasets

import pandas as pd
df = pd.read_csv("output_new.csv")
df2 = pd.read_csv("complete.csv")
data = df['len_data'].to_numpy()
data2 = df['len_40'].to_numpy()
data3 = df2['dist_search'].to_numpy()
```

```python
#HMM Model
gm = hmm.GaussianHMM(n_components=3)
gm.fit(data.reshape(-1, 1))
states = gm.predict(data.reshape(-1, 1))
#Plot
color_dict = {0:"r",1:"g",2:"b"}
color_array = [color_dict[i] for i in states]
plt.scatter(range(len(data)), data, c=color_array)
plt.title("HMM Model")
```
