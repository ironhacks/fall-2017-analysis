```python
import pandas as pd

df1 = pd.read_csv("final_visual.csv")
df1 = df1[df1['type'] == 'complete']
df2 = pd.read_csv("raw_data.csv")
df3 = pd.read_csv("novel.csv")
```

```python
df1.rename(columns={'dist_search': 'dist_search_vis', 'link1_change': 'username'}, inplace=True)
df1 = df1[['username', 'stage', 'dist_search_vis', 'type']]
df1
```

```python
df2.rename(columns={'dist_search': 'dist_search_data', 'user1': 'username'}, inplace=True)
df2 = df2[['username', 'stage', 'dist_search_data', 'type']]
df2
```

```python
df3.rename(columns={'dist_search': 'dist_search_novel', 'user1': 'username'}, inplace=True)
df3['type'] = df3['type'].map({'data': 'novel'})
df3 = df3[['username', 'stage', 'dist_search_novel', 'type']]
df3
```

```python
temp = pd.merge(df1, df2, how = "outer")
df = pd.merge(df3, temp, how = "outer")
df = df.fillna(0)
df.to_csv("cross_section.csv")
```

```python
df
```

```python
dfa = df[['username', 'dist_search_vis', 'stage', 'type']]
dfb = df[['username', 'dist_search_data', 'stage', 'type']]
dfc = df[['username', 'dist_search_novel', 'stage', 'type']]
```

```python
dfa.rename(columns={'dist_search_vis': 'dist_search'}, inplace=True)
dfb.rename(columns={'dist_search_data': 'dist_search'}, inplace=True)
dfc.rename(columns={'dist_search_novel': 'dist_search'}, inplace=True)
```

```python
df = pd.concat([dfa, dfb, dfc], ignore_index=True)
df['dist_search'] = df['dist_search'].replace([0], 0.01)
df['dist_search'] = df['dist_search'].replace([1], 0.99)
df
```

```python
df.to_csv("panel_data.csv")
```
