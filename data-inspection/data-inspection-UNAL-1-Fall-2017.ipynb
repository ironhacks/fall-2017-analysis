## This notebook is to do a preliminary exploration on the data quality generated from SSIM to determine the optimum research question. 

```python
import pandas as pd
import seaborn as sns

# UNAL 0 Phase 4 - 5 Raw Results with Colors

df = pd.read_csv("../data/Unal-0-Phase-4-_Unal-0-Phase-5–raw-results.csv")
df_maps = df[df['comparison_type'] == 'maps']
df_charts = df[df['comparison_type'] == 'charts']
df_complete = df[df['comparison_type'] == 'complete']
df_simple = df[df['comparison_type'] == 'simple']

print (pd.concat([
    df_maps.describe().reset_index().add_suffix('_maps'),
    df_charts.describe().reset_index().add_suffix('_charts'),
    df_complete.describe().reset_index().add_suffix('_complete'),
    df_simple.describe().reset_index().add_suffix('_simple')
  ], axis=1))
```

## Construct panel data for UNAL 1 with colors 

```python
df_45 = pd.read_csv("../data/Unal-1-Phase-4-_Unal-1-Phase-5–raw-results.csv")
df_34 = pd.read_csv("../data/Unal-1-Phase-3-_Unal-1-Phase-4–raw-results.csv")
df_23 = pd.read_csv("../data/Unal-1-Phase-2-_Unal-1-Phase-3–raw-results.csv")
df_12 = pd.read_csv("../data/Unal-1-Phase-1-_Unal-1-Phase-2–raw-results.csv")
df_12['stage'] = 1
df_23['stage'] = 2
df_34['stage'] = 3
df_45['stage'] = 4
```

```python
df_color = pd.concat([df_12, df_23, df_34, df_45], ignore_index=True)
df_color
```

```python
ax = sns.lineplot(data=df_color, x='stage', y='ssim', hue='comparison_type')
ax.legend(loc='center right', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

## Construct panel data for UNAL 1 with no colors

```python
df_45 = pd.read_csv("../data/Unal-1-Phase-4-_Unal-1-Phase-5–no-color-raw-results.csv")
df_34 = pd.read_csv("../data/Unal-1-Phase-3-_Unal-1-Phase-4–no-color-raw-results.csv")
df_23 = pd.read_csv("../data/Unal-1-Phase-2-_Unal-1-Phase-3–no-color-raw-results.csv")
df_12 = pd.read_csv("../data/Unal-1-Phase-1-_Unal-1-Phase-2–no-color-raw-results.csv")
df_12['stage'] = 1
df_23['stage'] = 2
df_34['stage'] = 3
df_45['stage'] = 4
```

```python
df_no_color = pd.concat([df_12, df_23, df_34, df_45], ignore_index=True)
df_no_color
```

```python
ax = sns.lineplot(data=df_no_color, x='stage', y='ssim', hue='comparison_type')
ax.legend(loc='center right', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

## Individual Exploration for UNAL 1

```python
df_no_color['link1_change'] = df_no_color['link1'].str.split('/', expand=True)[4]
df_no_color
```

```python
df_color['link1_change'] = df_color['link1'].str.split('/', expand=True)[4]
df_color['link2_change'] = df_color['link2'].str.split('/', expand=True)[4]
df_color
```

```python
df_color_exp = df_color[df_color['link1_change'] == df_color['link2_change']]
df_color_exp['ssim_exp'] = 100 - df_color_exp['ssim']
```

```python
ax = sns.lineplot(data=df_color_exp, x='stage', y='ssim_exp', hue='comparison_type')
ax.legend(loc='center right', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

```python
df_no_color['link1_change'] = df_no_color['link1'].str.split('/', expand=True)[4]
df_no_color['link2_change'] = df_no_color['link2'].str.split('/', expand=True)[4]
df_no_color_exp = df_no_color[df_no_color['link1_change'] == df_no_color['link2_change']]
df_no_color_exp['ssim_exp'] = 100 - df_no_color_exp['ssim']
```

```python
ax = sns.lineplot(data=df_no_color_exp, x='stage', y='ssim_exp', hue='comparison_type')
ax.legend(loc='center right', bbox_to_anchor=(1.25, 0.5), ncol=1)
```
