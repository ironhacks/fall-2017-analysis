## Descriptive Analysis for UNAL 0 Fall 2017

```python
import pandas as pd
import seaborn as sns

# UNAL 0 Phase 4 - 5 Raw Results with Colors

df = pd.read_csv("../data/Unal-0-Phase-4-_Unal-0-Phase-5–raw-results.csv")
df_maps = df[df['comparison_type'] == 'maps']
df_charts = df[df['comparison_type'] == 'charts']
df_complete = df[df['comparison_type'] == 'complete']
df_simple = df[df['comparison_type'] == 'simple']

print (pd.concat([
    df_maps.describe().reset_index().add_suffix('_maps'),
    df_charts.describe().reset_index().add_suffix('_charts'),
    df_complete.describe().reset_index().add_suffix('_complete'),
    df_simple.describe().reset_index().add_suffix('_simple')
  ], axis=1))
```

## Construct panel data for UNAL 2 with colors 

```python
df_45 = pd.read_csv("../data/Unal-0-Phase-4-_Unal-0-Phase-5–raw-results.csv")
df_34 = pd.read_csv("../data/Unal-0-Phase-3-_Unal-0-Phase-4–raw-results.csv")
df_23 = pd.read_csv("../data/Unal-0-Phase-2-_Unal-0-Phase-3–raw-results.csv")
df_12 = pd.read_csv("../data/Unal-0-Phase-1-_Unal-0-Phase-2–raw-results.csv")
df_12['stage'] = 1
df_23['stage'] = 2
df_34['stage'] = 3
df_45['stage'] = 4
```

```python
df_color = pd.concat([df_12, df_23, df_34, df_45], ignore_index=True)
df_color
```

```python
ax = sns.lineplot(data=df_color, x='stage', y='ssim', hue='comparison_type')
ax.legend(loc='center right', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

```python
import networkx as nx
import pandas as pd

df_color_stage1 = df_color[df_color['stage'] == 1]
df_color_stage1 = df_color_stage1[df_color_stage1['link1_change'] != df_color_stage1['link2_change']]
df_color_stage1 = df_color_stage1[['link1_change', 'link2_change', 'ssim']]

# Create Graph
G = nx.from_pandas_edgelist(
    df_color_stage1,
    source='link2_change',
    target='link1_change',
    edge_attr='ssim'
)

# Build adjacency matrix
adjacency_df = pd.DataFrame(
    nx.adjacency_matrix(G, weight='ssim').todense(),
    index=G.nodes,
    columns=G.nodes
)

adjacency_df.to_csv('stage1_unal0.csv')
```
